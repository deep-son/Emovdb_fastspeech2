```
dataset_path = "/content/EmoV-DB-sorted" 
final_path = "/content/emovdb"
```

```
#make speaker id folders at final_path
import os
import shutil

count = 1000
nodir = len(next(os.walk(dataset_path))[1])
for i in range(nodir):
  os.mkdir(final_path+"/"+str(count))
  count+=1
```

```
#copy from dataset to newly created folders
count = 1000
for filename in os.listdir(dataset_path): 
  src = dataset_path+"/" + filename 
  dst = dataset_path+"/" + str(count)
  os.rename(src, dst) 
  for folders, subfolders, filenames in os.walk(dst):
    for filename1 in filenames:
      fsrc = folders+ "/" + filename1
      fdst = final_path + "/" +str(count)
      shutil.copy(fsrc,fdst)
      print("Copied {} to {}".format(fsrc,fdst))
  count += 1 
```

```
#rename and remove out of convention files
#Neutral:000001, Amused:000002, sleepy:000003, disgust:000004, anger:000005
name_map ={"Neutral":"000001","neutral":"000001","amused":"000002","sleepiness":"000003","sleepy":"000003","disgust":"000004","Disgust":"000004","anger":"000005"}
bad_files=0
for folders, subfolders, filenames in os.walk(final_path):
  for filename1 in filenames:
    folder = folders.split("/")[-1]
    fname = filename1.split("_")
    try:
      k = int(fname[-1].split(".")[0])
      new_name  = folder + "_" + name_map[fname[0]]+ "_" + "000000" + "_00" + fname[2]
      src = os.path.join(folders,filename1)
      dst = os.path.join(folders,new_name)
      os.rename(src, dst)
    except:
      bad_files+=1
      pass
         
print(bad_files)
```

```
from scipy.io import wavfile
import scipy.signal as sps
for folders, subfolders, filenames in os.walk(final_path):
  for filename in filenames:
    path = os.path.join(folders,filename)
    new_rate = 24000
    sampling_rate, data = wavfile.read(path)
    number_of_samples = round(len(data) * float(new_rate) / sampling_rate)
    data = sps.resample(data, number_of_samples)
    os.remove(path)
    print("Removed 16k {}".format(path))
    os.chdir(folders)
    wavfile.write(path.split('/')[-1], new_rate, data)
    print("Added 24k   {}".format(path))
```

```
import os
import random
import shutil
import sys

libri_path = "/content/LibriTTS" # absolute path to TensorFlowTTS.
dataset_path = "/content/libritts" # Change to your paths. This is a output of re-format dataset.
subset = "train-clean-100"
```

```
with open(os.path.join(libri_path, "SPEAKERS.txt")) as f:
    data = f.readlines()
    
dataset_info = {}
max_speakers = 20  # Max number of speakers to train on
min_len = 20  # Min len of speaker narration time
max_file_len = 11  # max audio file lenght
min_file_len = 2  # min audio file lenght
```

```
possible_dataset = [i.split("|") for i in data[12:] if i.split("|")[2].strip() == subset and float(i.split("|")[3].strip()) >= min_len]
ids = [i[0].strip() for i in possible_dataset]
import soundfile as sf
possible_map = {}
subset_path = os.path.join(libri_path, subset)
for i in os.listdir(subset_path):
    if i in ids:
        id_path = os.path.join(subset_path, i)
        id_dur = 0
        id_included = []
        
        for k in os.listdir(id_path):
            for j in os.listdir(os.path.join(id_path, k)):
                if ".wav" in j:
                    f_path = os.path.join(id_path, k, j)
                    sf_file =  sf.SoundFile(f_path)
                    dur = len(sf_file) / sf_file.samplerate
                    if max_file_len < dur or dur < min_file_len:
                        continue
                    else:
                        id_included.append(f_path)
                        id_dur += dur
        possible_map[i] = {"dur": id_dur, "included": id_included}
poss_speakers = {k: v["included"] for k, v in possible_map.items() if v["dur"]/60 >= min_len}
```

```
to_move = list(poss_speakers.keys())
random.shuffle(to_move)
to_move = to_move[:max_speakers]
```

```
for sp_id, v in poss_speakers.items():
    if sp_id in to_move:
        for j in v:
            f_name = j.split(os.path.sep)[-1]
            text_f_name = f_name.split(".wav")[0] + ".txt"
            os.makedirs(os.path.join(dataset_path, sp_id), exist_ok=True)
            shutil.copy(j, os.path.join(dataset_path, sp_id, f_name))
            shutil.copy(j.replace(".wav", ".normalized.txt"), os.path.join(dataset_path, sp_id, text_f_name))
```
